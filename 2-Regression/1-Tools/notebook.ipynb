importing matplotlib, numpy and you are importing datasets, linear_model and model_selection from sklearn. model_selection is used for splitting data into training and test sets.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

The built-in diabetes dataset includes 442 samples of data around diabetes, with 10 feature variables, some of which include:

1. age: age in years
2. bmi: body mass index
3. bp: average blood pressure
4. s1 tc: T-Cells (a type of white blood cells)

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

```python
X = X[:, 2]
X = X.reshape((-1, 1))
```

```python
print(X)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color = 'black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
